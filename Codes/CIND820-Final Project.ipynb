```python
#To replace 'Not Available' values with blank spaces (empty strings) in numerical columns

# List of columns to process
columns_to_replace = ['WN_Sit_Elc_Int1','WN_Sit_Elc_Int2','WN_Sit_Gas_Int1','WN_Sit_Gas_Int2','WN_Sit_Gas_Int3',	'All_Water_Int1','All_Water_Int2','Ind_Water_Int1','Ind_Water_Int2','Site_EUI1','Site_EUI2','Source_EUI1','Source_EUI2','WN_Site_EUI1','WN_Site_EUI2',	'WN_Source_EUI1','WN_Source_EUI2',	'GHG_Emiss_Int1','GHG_Emiss_Int2','Ener_Star_Score']

# Replace 'Not Available' with blank (empty string) across specified columns
df[columns_to_replace] = df[columns_to_replace].replace('Not Available', '')
```

```python
# List of columns to convert from Object into numeric
columns_to_convert = ['WN_Sit_Elc_Int1','WN_Sit_Elc_Int2','WN_Sit_Gas_Int1','WN_Sit_Gas_Int2','WN_Sit_Gas_Int3',	'All_Water_Int1','All_Water_Int2','Ind_Water_Int1','Ind_Water_Int2','Site_EUI1','Site_EUI2','Source_EUI1','Source_EUI2','WN_Site_EUI1','WN_Site_EUI2',	'WN_Source_EUI1','WN_Source_EUI2',	'GHG_Emiss_Int1','GHG_Emiss_Int2']

# Convert specified columns to numeric, coercing errors into NaN
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# Convert specified columns to Integer, coercing errors into NaN
df['Ener_Star_Score'] = df['Ener_Star_Score'].fillna(0).astype(int)
```

```python
# List of columns to convert from Object into category
columns_to_convert = ['PrimPropTypCalc','PrimPropTypSelf','Largest_PropTyp','All_Prop_Types','Calculated with new source factors (Y/N)']

# Convert specified columns to cagegory
df[columns_to_convert] = df[columns_to_convert].astype('category')

df.dtypes
```
